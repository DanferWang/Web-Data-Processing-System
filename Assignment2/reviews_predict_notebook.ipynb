```
from datetime import datetime, timedelta

from tensorflow.keras.models import load_model
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder

import time
import pandas as pd
import pickle
```

```
# load the tokenizer
with open("/content/drive/MyDrive/ColabNotebooks/wdps/tokenizer.pkl","rb") as file:
  tokenizer = pickle.load(file)

# load the model
model = load_model("/content/drive/MyDrive/ColabNotebooks/wdps/model.h5")
```

```
def predict(text):
    # Tokenize text
    x_test = pad_sequences(tokenizer.texts_to_sequences([text]), maxlen=320)
    # Predict
    score = model.predict([x_test])[0]
    return round(float(score), 4)
```

```
test_text = "good"
predict(test_text)
```
